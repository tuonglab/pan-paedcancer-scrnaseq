#### 1. Plot the pie chars

```python
import matplotlib
import matplotlib.pyplot as plt

FONTTYPE = 42

# set the font so no weird lines and boxes are made during export
matplotlib.rcParams["pdf.fonttype"] = FONTTYPE
matplotlib.rcParams["ps.fonttype"] = FONTTYPE

palette = [
    "#4E79A7",
    "#F28E2B",
    "#E15759",
    "#76B7B2",
    "#59A14F",
    "#EDC948",
    "#B07AA1",
    "#FF9DA7",
    "#9C755F",
    "#BAB0AC",
]
```

```python
repo_dict = {
    "Gene Expression Omnibus (GEO)": 27,
    "European Genome-phenome Archive (EGA)": 7,
    "EGA & GEO": 2,
    "EGA & European Nucleotide Archive (ENA)": 2,
    "Genome Sequence Archive (GSA)": 2,
    "BioProject": 1,
    "Synapse": 1,
    "Unknown": 5,
}

# Extract labels and values
repositories = list(repo_dict.keys())
dataset_counts = list(repo_dict.values())
total_datasets = sum(dataset_counts)

# Create a list of labels showing the exact count and percentage
labels_with_counts = [f'{count} ({count/total_datasets*100:.1f}%)' for repo, count in zip(repositories, dataset_counts)]

# Plot the pie chart
plt.figure(figsize=(7, 5))
plt.pie(
    dataset_counts,
    labels=labels_with_counts,
    startangle=90,
    counterclock=False,
    colors=palette[:len(repositories)-1] + [palette[-1]],
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Repository')
plt.legend(repositories, loc='upper right', bbox_to_anchor=(2.1, 1))
plt.savefig("/scratch/user/s4543064/Xiaohan_Summer_Research/figures/Fig2B_repo.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

#### 2. Plot the UMAP

```python
import scanpy as sc
```

```python
adata = sc.read_h5ad("/scratch/user/s4543064/Xiaohan_Summer_Research/write/all_cancer_meta_anndata_scVI_pca_clean.h5ad")
adata
```

```python
# Update the major cancer type names
cancer_names = {
    'leukemia': 'Leukemia', 
    'healthy_donor': 'Healthy donor', 
    'brain_tumor': 'CNS tumour', 
    'bone_cancer': 'Bone cancer', 
    'rhabdomyosarcoma': 'Sarcoma', 
    'neuroblastoma': 'Peripheral neuroblastic tumour', 
    'retinoblastoma': 'Retinoblastoma', 
    'kidney_cancer': 'Kidney cancer'
}

adata.obs['major_cancer_type_order'] = adata.obs['major_cancer_type_order'].replace(cancer_names)
adata.obs["major_cancer_type_order"].value_counts()
```

```python
import pandas as pd

# Define the desired order
desired_order = [
    'Leukemia', 
    'CNS tumour', 
    'Kidney cancer', 
    'Peripheral neuroblastic tumour', 
    'Retinoblastoma', 
    'Sarcoma', 
    'Bone cancer', 
    'Healthy donor'
]

# Convert the column to a categorical type with the specified order
adata.obs['major_cancer_type_order'] = pd.Categorical(
    adata.obs['major_cancer_type_order'], 
    categories=desired_order, 
    ordered=True
)

# Check if it has been ordered
print(adata.obs['major_cancer_type_order'])
```

```python
patientnum_palette = [
    "#4E79A7",
    "#E15759",
    "#76B7B2",
    "#EDC948",
    "#B07AA1",
    "#FF9DA7",
    "#9C755F",
    "#D7B5A6"
]
```

```python
sc.settings.figdir = '/scratch/user/s4543064/Xiaohan_Summer_Research/figures/'
sc.settings.vector_friendly = False

sc.pl.umap(adata, color='major_cancer_type_order', frameon=False, palette=patientnum_palette,
           save='Fig3B_umap.pdf')
```
